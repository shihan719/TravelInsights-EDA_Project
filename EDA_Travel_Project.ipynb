<span style="font-family:Calibri"><span style="color:pink">
>How does trip duration vary throughout the year ?</span>

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your DataFrame is named 'travel_data'
# Convert 'Start date' and 'End date' columns to datetime
Travel['Start date'] = pd.to_datetime(Travel['Start date'])
Travel['End date'] = pd.to_datetime(Travel['End date'])

# Calculate trip duration in days
Travel['Trip duration'] = (Travel['End date'] - Travel['Start date']).dt.days

# Group by month and calculate average trip duration
monthly_avg_duration = Travel.groupby(Travel['Start date'].dt.month)['Trip duration'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_duration.index, monthly_avg_duration, marker='o', linestyle='-', color='b')
plt.title('Average Trip Duration Throughout the Year')
plt.xlabel('Month')
plt.ylabel('Average Trip Duration (days)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

<span style="font-family:Calibri"><span style="color:pink">
>Is there a correlation between the booking lead time and accommodation cost ?</span>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'travel_data'
# Ensure 'Start date' is in datetime format
Travel['Start date'] = pd.to_datetime(Travel['Start date'])

# Calculate booking lead time
Travel['Lead time'] = (Travel['Start date'] - pd.to_datetime('today')).dt.days

# Scatter plot
plt.scatter(Travel['Lead time'], Travel['Accommodation cost'])
plt.xlabel('Booking Lead Time (days)')
plt.ylabel('Accommodation Cost')
plt.title('Correlation between Booking Lead Time and Accommodation Cost')
plt.show()

# Calculate correlation coefficient
correlation_coefficient = Travel['Lead time'].corr(Travel['Accommodation cost'])
print(f"Correlation Coefficient: {correlation_coefficient}")
```

<span style="font-family:Calibri"><span style="color:pink">
>Create a heatmap to show the busiest months for travel across all destinations.<span>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming your travel data is stored in a DataFrame named 'travel_data'
# Make sure your 'Start date' column is in datetime format

# Convert 'Start date' to datetime if it's not already
Travel['Start date'] = pd.to_datetime(Travel['Start date'])

# Extract month and year from the 'Start date'
Travel['Month'] = Travel['Start date'].dt.month

# Create a pivot table with destination as index, month as columns, and count of trips as values
heatmap_data = pd.pivot_table(Travel, values='Trip ID', index='Destination', columns='Month', aggfunc='count', fill_value=0)

# Set up the heatmap using seaborn
plt.figure(figsize=(14, 14))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', cbar_kws={'label': 'Number of Trips'})
plt.title('Busiest Months for Travel Across Destinations')
plt.xlabel('Month')
plt.ylabel('Destination')

plt.show()
```

<span style="font-family:Calibri"><span style="color:pink">
>Use a grouped bar chart to compare the distribution of nationalities across destinations.</span>

```python
import matplotlib.pyplot as plt
import pandas as pd

Travel_grpbarchart = pd.read_excel(file1, sheet_name='Travel details dataset', header=0)

# Extract relevant columns
nationality_data = Travel_grpbarchart[['Destination', 'Traveler nationality']]

# Group by Destination and Traveler nationality, count occurrences
grouped_data = nationality_data.groupby(['Destination', 'Traveler nationality']).size().unstack()

# Plotting the grouped bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Nationalities Across Destinations')
plt.xlabel('Destination')
plt.ylabel('Number of Travelers')
plt.legend(title='Nationality', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```
